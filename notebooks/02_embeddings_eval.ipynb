# 02 — Embedding evaluation
Sanity-check the text embedder: similar sentences should cluster, dissimilar should not.

```python
import os, sys
ROOT = os.path.abspath('..')
for sub in ('packages/core', 'packages/ml'):
    p = os.path.join(ROOT, sub)
    if p not in sys.path: sys.path.insert(0, p)
import numpy as np
from echomind_ml.embed_text import TextEmbedder
```

```python
emb = TextEmbedder()
sentences = [
    'A creature wanders the icy glacier alone.',
    'On the frozen mountain, a being walked solitary.',
    'Elizabeth wrote a polite letter to her cousin.',
    'The dance lasted until well past midnight.',
]
X = emb.embed(sentences)
sim = X @ X.T
print(np.round(sim, 3))
```

## Interpretation
Sentences 0 and 1 (both about wandering in icy terrain) should be > 0.6. Sentences 2 and 3 should be lower vs 0 and 1.

```python
assert sim[0, 1] > sim[0, 2], 'expected the glacier pair to be closer than glacier vs letter'
print('OK')
```
